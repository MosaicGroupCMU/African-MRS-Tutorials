# Solution

```python
# install libraries for numerical calculations, plotting, and iterative functions
import numpy as np
import matplotlib.pyplot as plt
import tqdm

# discretize space
dx = 1
x = np.arange(-60, 61, dx)
M = len(x)


# set harmonic potential
k = 1e-4
v = k / 2 * x**2

# set square potential
# v0 = 10
# a = 60
# v = v0 * (x < -a/2) | (x > a/2)

# set number of energy levels to plot
N = 1

# set coefficients of Hamiltonian
H = np.zeros((M, M))
for m in range(M):
    H[m, m] = 2 / dx**2 + v[m]
for m in range(M-1):
    H[m, m+1] = -1 / dx**2
    H[m+1, m] = -1 / dx**2

lmda, psi = np.linalg.eig(H)
bottom_k = np.argsort(lmda)[:N]

# plot the result
scaling = 0.02
plt.plot(x, v, color='black')
_ = plt.plot(x, psi[:, bottom_k].real * scaling + lmda[bottom_k])
plt.ylim([0, 0.1])
plt.xlabel('x (a.u)')
plt.ylabel('energy (a.u.)')
plt.show()
```
